REMOVING 'NEW', '-', '/5' FROM RATE IN CLEANING PROCESS


np.nan is used to replace - and new with nan because there are so many methods to handle the null values


```python
def clean_rate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate'] = df['rate'].apply(clean_rate)
df['rate'].head()
```

HANDLE THE NULL VALUES (BY FILLING ALL THE NULL VALUES BY TAKING MEAN OF ALL THE RATE VALUES)


```python
from numpy import mean
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].unique()
```

```python
location = df['location'].value_counts(ascending=True)
location_less_200 = location[location<200]
location_less_200
def less(value):
    if value in location_less_200:
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(less)
df['location'].value_counts(ascending=True)
df.head(20)
```

```python
#CLEANING REST TYPE COLUMN
df['rest_type'].value_counts()
rest_type = df['rest_type'].value_counts(ascending= True)
rest_type
rest_type_lessthan_1000 = rest_type[rest_type<1000]
rest_type_lessthan_1000
def other(value):
    if value in rest_type_lessthan_1000:
        return 'other'
    else:
        return value
df['rest_type']  = df['rest_type'].apply(other)
df.head(5)
```

```python
df['cuisines'].value_counts(ascending=True)
cusines = df['cuisines'].value_counts(ascending=True)
cusines_less_300 = cusines[cusines<300]
cusines_less_300
```

```python
def cus(value):
    if value in cusines_less_300:
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(cus)
df['cuisines'].value_counts(ascending=True)    
df['Type'].value_counts()
```
